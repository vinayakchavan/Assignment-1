```python
# question 1 , day 1 data visualization using python 
x=np.arange(0,10)
y=x*x
plt.title('Line plot', color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.plot(x,y,'r*', linestyle='dashed',Linewidth=2, marksize=12)
plt.plot(x,y, 'ro--')
```

```python
# plotting in 2x2 plane in 1 , 2, 3, 4 section
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
# question 2
# day
x=[1,2,3,4,5,6,7] 
# sales
y=[160,150,140,145,175,165,180]
plt.title('sales data', color='blue')
plt.xlabel('days')
plt.ylabel('sales')
plt.plot(x,y)
```

```python
# question 3
# day
days=[1,2,3,4,5,6,7] 
# sales 1
sales_1=[160,150,140,145,175,165,180]
# sales 2
sales_2=[70,90,160,150,140,145,175]
plt.title('sales data', color='blue')
plt.subplot(3,3,1)
plt.xlabel('days')
plt.ylabel('sales 1')
plt.plot(days,sales_1)
plt.subplot(3,3,3)
plt.xlabel('days')
plt.ylabel('sales 2')
plt.plot(days,sales_2)
```
